```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/drive/MyDrive/Salary Data.csv")
print(df.head())
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace = True)
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df.dropna(axis = 1)
```

```python
df_encoded=pd.get_dummies(df,drop_first=True)
```

```python
print(df_encoded.head())
```

```python
scaler = StandardScaler()
data_scaled = df.copy()
columns_to_normalize = ['Age', 'Years of Experience', 'Salary']
data_scaled[columns_to_normalize] = scaler.fit_transform(data_scaled[columns_to_normalize])
```

```python
# Split the data into training and testing sets (e.g., 80% training, 20% testing).

from sklearn.model_selection import train_test_split

X = df_encoded.drop('Salary', axis=1)
y = df_encoded['Salary']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
linear_model=LinearRegression()
linear_model.fit(X_train,y_train)
```

```python
y_pred=linear_model.predict(X_test)
# 2. Model Evaluation:
# Evaluate the performance of the model using metrics like:
# Mean Absolute Error (MAE)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mean_absolute_error(y_test, y_pred)
```
